```python
import numpy as np
```

```python
single = np.array([array,])
```

```python
model.predict(single)
```

```python
%%timeit
model.predict(single)
```

25500 microseconds

Now make a list of 10000 random boards

```python
array_list = []
for i in range(10000):
    b.randomize(30)
    array_list.append(b.get_flipped_3d_array())
```

```python
array_list[0]
```

```python
array_list[1]
```

```python
long_input = np.array(array_list)
```

```python
%%timeit
model.predict(long_input)
```

121 microseconds

Hmm, so it takes 1.21 sec to do 10000. 
So the effective speed is 0.1210 millisecs

This is ~250 times faster than the 25 millisecs to do one board

```python
%%timeit
model(single)
```

```python
%%timeit
model.predict(shorter_input)
```

440 microseconds
