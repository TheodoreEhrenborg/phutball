```python
import pickle
```

```python
boards = []
with open('2021-08-04-3ply-boards5.pickle', 'rb') as f:
    while True:
        boards.append(pickle.load(f))
```

```python
len(boards)
```

```python
for (b,s) in boards:
    b.pretty_print_details()
    print("Score:", s)
```
